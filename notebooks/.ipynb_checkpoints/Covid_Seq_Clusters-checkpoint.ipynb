# 1. Data Cleaning

Import data and look for missing values and outliers.

Data Observations:

- Most sequences have random sections of missing values
- Both ends of the sequences have ragged alignment

Cleaning Approach:

- For positions with <5% invalid symbols ('-','.','N'), substitute the most common nucleotide value.
- For positions with >5% invalid symbols, omit from analysis.

---

Import Packages


```python
import matplotlib.pyplot as plt
import numpy as np
from numba import jit
from sklearn import cluster, neighbors
from utils import kmeans as jm_km
from scipy import sparse
```

Read FASTA file.


```python
check_performance = True
raw_seq_file = "../data/Modified_Sequences.fasta"

file = open(raw_seq_file, "r")
lines = file.readlines()
file.close()
```

Parse FASTA file. Convert ASCII characters to int16 values.


```python
def parse_sequences(rows):
    seq_names = np.array([name.strip() for name in rows[::2]])
    seqs = np.array([list(mers.strip()) for mers in rows[1::2]], dtype="S1")
    seqs = np.array(seqs).view(np.int8)
    return seqs, seq_names


sequences, _ = parse_sequences(lines)
sequences[:5, 100:110]
```

Look at frequency of the various symbols and nucleotides.


```python
unique, frequency = np.unique(sequences, return_counts=True)
print(unique)
print([chr(c) for c in list(unique)])
print(frequency)
```

Visualize Sequences.

See saved image for more detail.


```python
plt.figure(figsize=(30, 30))
plt.imshow(sequences, cmap="gray")
plt.imsave("../plots/sequences.png", sequences, cmap="gray")
```

Only look at initial 500 base-pairs.


```python
plt.figure(figsize=(5, 5))
plt.imshow(sequences[:, :500], cmap="gray")
plt.imsave("../plots/sequences_head.png", sequences[:, :500], cmap="gray")
```

Only look at last 500 base pairs.


```python
plt.figure(figsize=(5, 5))
plt.imshow(sequences[:, -500:], cmap="gray")
plt.imsave("../plots/sequences_tail.png", sequences[:, -500:], cmap="gray")
```

For each position, check if <5% of symbols are invalid.

- If so, impute the most common symbol
- Else remove position from analysis.


```python
n, m = sequences.shape
invalid_chars = [45, 46, 78]  # 45='-', 46='.', 78='N"
valid_threshold = 0.25

clean_seqs = sequences.copy()
keep_mask = np.full(m, True)

for a in range(m):
    mer = clean_seqs[:, a]

    invalid_mask = np.in1d(mer, invalid_chars)
    valid_ratio = np.sum(invalid_mask) / n

    if valid_ratio < valid_threshold:
        keep_mask[a] = True

        vals, counts = np.unique(mer, return_counts=True)
        mode_index = np.argwhere(counts == np.max(counts))
        mode_val = vals[mode_index][0][0]

        mer[invalid_mask] = mode_val

    else:
        keep_mask[a] = False

clean_seqs = clean_seqs[:, keep_mask]
print(clean_seqs.shape)
clean_seqs[:5, :10]
```

Check which symbols and associated frequency, post cleaning.


```python
unique, frequency = np.unique(clean_seqs, return_counts=True)
print(unique)
print([chr(c) for c in list(unique)])
print(frequency)
```

Visualize Sequences, post-cleaning


```python
plt.figure(figsize=(30, 30))
plt.imshow(clean_seqs, cmap="gray")
plt.imsave("../plots/sequences_clean.png", clean_seqs, cmap="gray")
```

Only look at first 500 base pairs, post-cleaning.


```python
plt.figure(figsize=(5, 5))
plt.imshow(clean_seqs[:, :500], cmap="gray")
plt.imsave("../plots/sequences_clean_head.png", clean_seqs[:, :500], cmap="gray")
```

Only look at last 500 base pairs, post-cleaning.


```python
plt.figure(figsize=(5, 5))
plt.imshow(clean_seqs[:, -500:], cmap="gray")
plt.imsave("../plots/sequences_clean_tail.png", clean_seqs[:, -500:], cmap="gray")
```

# 2. Spectral Clustering

Here I'll implement spectral clustering of the Hamming distance matrix by building a graph.

Since the Hamming distance is pairwise, the matrix needs to be 'mapped' to a coordinate system prior to clustering.

Steps:

1. Distance metric used is the Hamming distance, which is used to determine the distance between sequence strings.
2. Build a k-nearest neighbors adjacency matrix $A$ using the Hamming matrix.
3. Will build a graph Laplacian matrix $L$ based on Adjacency matrix $A$ and Degree matrix $D$
   - Try both $L = D - A$ and $L = D^{-\frac{1}{2}} \cdot A \cdot D^{-\frac{1}{2}}$ methods
4. Find eigenpairs (spectum) and select $k$ smallest (or largest for $L = D^{-\frac{1}{2}} \cdot A \cdot D^{-\frac{1}{2}}$)
5. Use $k$-means to fit $k$ clusters to the selected $k$ eigenvectors.

---

Calculate Hamming Distance for all pairs.


```python
@jit(nopython=True)
def get_hamming_distance(seqs, base_id=0):
    model_org = seqs[base_id]
    ham_mask = np.not_equal(seqs, model_org)
    return np.sum(ham_mask, axis=1)


@jit(nopython=True)
def get_hamming_matrix(seqs):
    l = len(seqs)
    Dh = np.zeros((l, l), dtype="int16")
    for i in range(l):
        Dh[:, i] = get_hamming_distance(seqs, i)
    return Dh
```

Assign Hamming Distance to Adjacency Matrix $A$


```python
A = get_hamming_matrix(clean_seqs)
A[:10, :10]
```

Display adjacency matrix as an image.

Note: no gross patterns as were observed prior to cleaning data.


```python
plt.imshow(A, cmap="gray")
plt.imsave("../plots/distance_matrix.png", A, cmap="gray")
```

Plot histogram of _maximum_ distances across sequences


```python
A_max = np.max(A, axis=0)
_ = plt.hist(A_max, bins=50)
```

Plot histogram of _mean_ distances across sequences

A potential niave method would be to cluster the mean distances. However, this may be problematic for network adjacency data.


```python
A_mean = np.mean(A, axis=0)
_ = plt.hist(A_mean, bins=50)
```

Plot histogram of _standard deviation_ of distances across sequences


```python
A_std = np.std(A, axis=0)
_ = plt.hist(A_std, bins=50)
```

To improve performance of the clustering, reduce connections in graph to only incluide the $k$ nearest neighbors

Convert sparse matrix (CSC) to dense np.ndarray.


```python
k_nn = 75
knn_distance = neighbors.kneighbors_graph(
    A, n_neighbors=k_nn, include_self=False, mode="distance"
)
A_nn = np.asarray(sparse.csc_matrix.todense(knn_distance))
A_nn[:5, :5]
```

Calculate Degree matrix $D$

Alternatively, tried $D^{-\frac{1}{2}}$


```python
d = np.sum(A_nn, axis=1)
D = np.diag(d)
D[:5, :5]
# D_inv_root = np.diag(1/np.sqrt(d))
# D_inv_root[:5,:5]
```

Calculate the Graph Laplacian matrix $L$

Note: The $L = D - A$ method gives the best results. The $L = D^{-\frac{1}{2}} \cdot A \cdot D^{-\frac{1}{2}}$ method didn't perform as well.


```python
# L = D_inv_root @ A3 @ D_inv_root
L = D - A_nn
L[:5, :5]
```

Get eigenpairs.


```python
v, x = np.linalg.eig(L)
v = v.real
v_sort = np.sort(np.abs(v))
v_sort[:5]
```

Sort eigenvectors in ascending order based on eigenvalues.


```python
x = x.real
idx = np.argsort(np.abs(v))  # the index of eigenvalue sorted acsending
x_sort = x[:, idx]  # select the k largest eigenvectors
x_sort[:5, :5]
```

Function to visualize eigenvectors.


```python
def plot_eigs(i, j, v, labels=None):
    fig, ax = plt.subplots()
    plt.scatter((v[:, i]), (v[:, j]), c=labels)

    ax.set_xlabel(f"Eigenvector {i}")
    ax.set_ylabel(f"Eigenvector {j}")
    ax.set_title(f"Eigenvectors {i} & {j}")
    plt.show()
```

Look at a few combinations of eignvectors.


```python
for i, j in zip(range(0, 3, 1), range(1, 4, 1)):
    plot_eigs(i, j, x_sort)
```

Function to execute $k$-means

Note that spectral clustering uses two parameters, $n_{eig}$ eigenvectors to fit $k$ clusters. Typically these two values are equal, but adding dimensions makes the elbow diagram unstable, so $n_{eig}$ was selected that gives good definition between cluster inertia values.


```python
def run_kmeans(k, v, n_init):
    v_k = v[:, :k]
    kmeans = cluster.KMeans(n_clusters=k, n_init=n_init, random_state=42).fit(v_k)
    return kmeans
```

Function to plot elbow diagram.


```python
def elbow_diagram(norm_inertia, x):
    fig, ax = plt.subplots()
    plt.plot(x, norm_inertia)
    ax.set_ylabel("Inertia / k")
    ax.set_xlabel("Number of Clusters (k)")
    ax.set_title(f"Elbow Diagram")
    plt.show()
```

Execute $k$-means over a range of $k$ values and plot _Elbow Diagram_.


```python
k_min = 1  # 0-index
k_max = 7
norm_inertia = [
    run_kmeans(k, x_sort, n_init=25).inertia_ / k for k in range(k_min, k_max)
]
ct = [*range(k_min, k_max)]
elbow_diagram(norm_inertia, ct)
```

It appears that there $k=4$ is the best fit on the elbow diagrams.

# 4. Alternate Method: ISOMAP

Given Pairwise Dissimilarity between Data Points, Reconstruct a low-dimensional distance "map" that preserves distance.

Refer to paper by _Tenenbaum, Joshua B. A Global Geometric Framework for Nonlinear Dimensionality Reduction. \_Science_, 290:2319-2323, 2000.\_

Often used to find manifolds in high-dimension space, such as the classic 'swiss-roll' manifold.


```python
A[:5, :5]
```

Find the shortest path between nodes in k-Nearest Neighbor network


```python
from scipy.sparse.csgraph import shortest_path

k_nn = 165
knn_distance = neighbors.kneighbors_graph(
    A, n_neighbors=k_nn, include_self=False, mode="distance"
)
A_nn = np.asarray(sparse.csc_matrix.todense(knn_distance))

D2 = shortest_path(A_nn)
D2[:5, :5]
```

Calculate centering matrix $H$


```python
m = D2.shape[0]
H = np.eye(m) - 1 / m * np.ones((m, m))
H[:5, :5]
```

Center the adjacency matrix $D$ by $H$


```python
G = -1 / 2 * H @ np.square(D2) @ H
G[:5, :5]
```

Find the $n_eigs$ number of leading eigenvalues and eigenvectors


```python
n_eigs = 10
import scipy.sparse.linalg as ll

lam, u = ll.eigs(G, k=n_eigs)
lam = lam.real
u = u.real
lam[:5]
```

Calculate the reduced representation matrix $Z$


```python
Z_T = u @ np.diag(lam)
Z = Z_T.T
Z_T[:5, :5]
```

Run $k$-means clustering to find $k$ with best fit.


```python
k_min = 1  # 0-index
k_max = 15
norm_inertia = [run_kmeans(k, Z_T, n_init=25).inertia_ / k for k in range(k_min, k_max)]
ct = [*range(k_min, k_max)]
elbow_diagram(norm_inertia, ct)
```

$k=3$ Appears to fit the data best.

Get cluster labels:


```python
k = 5
km = cluster.KMeans(n_clusters=k).fit(Z_T[:, :n_eigs])
labels = km.labels_
```

Visualize labeled clusters

Note that


```python
i = 0
j = 1
fig, ax = plt.subplots()
plt.scatter((Z_T[:, i]), (Z_T[:, j]), c=labels)

ax.set_xlabel(f"Reduced Dimension {i}")
ax.set_ylabel(f"Reduced Dimension {j}")
ax.set_title(f"2D Reduced Representation")
plt.show()
```

## Conclusion

The MDS approach shows differentiation between clusters of data. Between 3 and 5 clusters colud be seleceted to fit the data.

Known genomic sequence data should be used to validate this method of representing the Hamming distance matrix.
